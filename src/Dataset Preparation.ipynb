```python
import pandas as pd
from sklearn.model_selection import StratifiedKFold

# Load train, test, and val dataframes
fold = 3

train_df = pd.read_csv(f'../dataset/10-fold_original/algocite_utilize_dataset_original_train_fold_{fold}.csv')
val_df = pd.read_csv(f'../dataset/10-fold_original/algocite_utilize_dataset_original_val_fold_{fold}.csv')
test_df = pd.read_csv(f'../dataset/10-fold_original/algocite_utilize_dataset_original_test_fold_{fold}.csv')

# Define the target column name
target_column = 'UTILIZE_LABELS'

# Create a StratifiedKFold object with 100 folds
skf = StratifiedKFold(n_splits=100, shuffle=True, random_state=42)

# Initialize a list to hold the indices for each fold
fold_indices = []

# Loop over the splits generated by the StratifiedKFold object
for fold_number, (train_index, test_index) in enumerate(skf.split(test_df, test_df[target_column])):
    # Add the indices for the current fold to the list
    fold_indices.append(test_index)

    # Print some information about the current fold
    print(f"Fold {fold_number}: Train size = {len(train_index)}, Test size = {len(test_index)}")

# Select the rows corresponding to the first fold
# fold1 = train_df.loc[fold_indices[0]]
```
